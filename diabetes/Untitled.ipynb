```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
col_lst = []
for col in df.columns:
    if (df[col].min() == 0) & (col != 'Outcome'):
        col_lst.append(col)

df[col_lst] = df[col_lst].replace(0, np.NaN)
```
